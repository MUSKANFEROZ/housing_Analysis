```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
print(df.head())
```

```python
#Explore & Clean Data

print(df.info())
print(df.describe())
```

```python
#Check and handle any missing values.

print(df.isnull().sum)
```

```python
print(df.columns)
```

```python
# Assuming columns: 'price', 'area', 'bedrooms'
cols = ['median_house_value', 'total_rooms', 'total_bedrooms']

#Compute Statistics
for col in cols:
    print(f"\nStatistics for {col.upper()}:")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Mode: {df[col].mode()[0]}")
    print(f"Variance: {df[col].var()}")
    print(f"Standard Deviation: {df[col].std()}")
    print(f"Skewness: {df[col].skew()}")
    print(f"Kurtosis: {df[col].kurt()}")
    print(f"Min: {df[col].min()}")
    print(f"Max: {df[col].max()}")
```

```python
# Visualization

plt.hist(df['median_house_value'], bins=20, color= 'skyblue' )
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.title('Distribution of Median House Value')
plt.show()
```

```python
plt.scatter(df['total_rooms'], df['median_house_value'], color= 'red')
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')
plt.title('Relationship between Total Rooms and Median House Value')
plt.show()
```
